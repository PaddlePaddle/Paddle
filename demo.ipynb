### demo1: 支持map-style数据集对应的DataPipes系列API

```python
import numpy as np
from paddle.io import Dataset
from paddledata.datapipes.map import *
```

```python
class RandomMapDataset(Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __getitem__(self, idx):
        data = self.x[idx]
        label = self.y[idx]
        return data, label
    
    def __len__(self):
        return len(self.x)

x = np.random.random([10, 3]).astype('float32')
y = np.random.randint(0, 2, (10, 1)).astype('int64')
map_dataset = RandomMapDataset(x, y)

for i in range(len(map_dataset)):
    print(map_dataset[i])
```

```python
def add_one(x):
    return x + 1

map_dataset_1 = MapMapper(map_func=add_one)(map_dataset)
for i in range(len(map_dataset_1)):
    print(map_dataset_1[i])
```

```python
map_dataset_2 = MapMapper(map_func=lambda x: x + 2)(map_dataset)
for i in range(len(map_dataset_2)):
    print(map_dataset_2[i])
```

```python
map_dataset_3 = MapFilter(predicate=lambda z: z[0][1] > 0.5)(map_dataset)
for i in range(len(map_dataset_3)):
    print(map_dataset_3[i])
```

```python
map_dataset_4 = MapShuffler()(map_dataset)
for i in range(len(map_dataset_4)):
    print(map_dataset_4[i])
```

```python
map_dataset_5 = MapBatcher(batch_size=4, drop_last=False)(map_dataset)
for i in range(len(map_dataset_5)):
    print(map_dataset_5[i])
```

```python
map_dataset_6 = MapBatcher(batch_size=4, drop_last=True)(map_dataset)
for i in range(len(map_dataset_6)):
    print(map_dataset_6[i])
```

```python
Pipelines = list()
Pipelines.append(MapMapper(map_func=lambda x: x + 1))
Pipelines.append(MapFilter(predicate=lambda x: x[1] > 0.5))
Pipelines.append(MapShuffler())
Pipelines.append(MapBatcher(batch_size=4, drop_last=False))
DataPipes = MapDataPipes(pipelines=Pipelines)

map_dataset_7 = DataPipes(map_dataset)
for i in range(len(map_dataset_7)):
    print(map_dataset_7[i])
```

### demo2: 支持iterable-style数据集对应的DataPipes系列API

```python
import numpy as np
from paddle.io import IterableDataset
from paddledata.datapipes.iter import *
```

```python
class RandomIterableDataset(IterableDataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __iter__(self):
        for i in range(len(self.x)):
            data = self.x[i]
            label = self.y[i]
            yield data, label

x = np.random.random([10, 3]).astype('float32')
y = np.random.randint(0, 2, (10, 1)).astype('int64')
iterable_dataset = RandomIterableDataset(x, y)

for data, label in iterable_dataset:
    print(data, label)
```

```python
def add_one(x):
    return x + 1

iterable_dataset_1 = IterableMapper(map_func=add_one)(iterable_dataset)
for data, label in iterable_dataset_1:
    print(data, label)
```

```python
iterable_dataset_2 = IterableMapper(map_func=lambda x: x + 2)(iterable_dataset)
for data, label in iterable_dataset_2:
    print(data, label)
```

```python
iterable_dataset_3 = IterableFilter(predicate=lambda z: z[0][1] > 0.5)(iterable_dataset)
for data, label in iterable_dataset_3:
    print(data, label)
```

```python
iterable_dataset_4 = IterableShuffler(replace=False)(iterable_dataset)
for data, label in iterable_dataset_4:
    print(data, label)
```

```python
iterable_dataset_5 = IterableBatcher(batch_size=4, drop_last=False)(iterable_dataset)
for data, label in iterable_dataset_5:
    print(data, label)
```

```python
Pipelines = list()
Pipelines.append(IterableMapper(map_func=lambda x: x + 1))
Pipelines.append(IterableFilter(predicate=lambda x: x[1] > 0.5))
Pipelines.append(IterableShuffler())
Pipelines.append(IterableBatcher(batch_size=4, drop_last=False))
DataPipes = IterableDataPipes(pipelines=Pipelines)

iterable_dataset_6 = DataPipes(iterable_dataset)
for data, label in iterable_dataset_6:
    print(data, label)
```

### demo4: 文本解码

```python
!pip install gensim
```

```python
from paddle.io import IterableDataset
from paddledata.text_decoding import *
import jieba
from gensim.models import word2vec
```

```python
corpus = read_txt(path='data/decompression/tar/Sina_News.txt')[:5]
print(corpus)
```

```python
sentences = []
for i, j in corpus:
    d = list(jieba.cut(i))
    sentences.append(d)

print(sentences)
```

```python
model = word2vec.Word2Vec(sentences=sentences,min_count=1,window=5,sg=1,hs=1)

embedding = []
for s in sentences:
    words = []
    for w in s:
        words.append(model.wv[w])
    embedding.append(words)

print(embedding)
```

```python
class MyDataset(IterableDataset):
    def __init__(self, paths):
        self.paths = paths
    
    def __iter__(self):
        for path in self.paths:
            yield path

paths = ['data/decompression/tar/Sina_News.txt']
example = MyDataset(paths)
for i in example:
    print(i)
```

```python
preparation = Decoding()(example)
for data, label in preparation:
    print(data, label)
```
