```python
import torch
import numpy as np
from scipy import special, misc
```

```python
def ref_i0e_grad(x, dout):
    gradx = special.i1e(x) - np.sign(x) * special.i0e(x)
    return dout * gradx
```

```python
case = [0.]
a = torch.tensor(case, requires_grad=True, dtype=torch.float64)
out = torch.i0(a) / torch.exp(a)

out.backward()

print(a.grad)

a = np.array(case).astype("float64")
out = ref_i0e_grad(a, 1 / a.size)
out
```
